```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image_path = "../videos/output_videos/cropped_image.jpg"
image = cv2.imread(filename=image_path)
image = cv2.cvtColor(src=image, code=cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

# TAKE THE TOP HALF OF THE IMAGE

```python
top_half = image[0: int(image.shape[0]/2), :]
plt.imshow(top_half)
plt.show()
```

# CLUSTER THE IMAGE INTO TWO

```python
# Reshape the image
image_2d = top_half.reshape(-1, 3)

# perform k-means clustering with two clusters
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(image_2d)

# get cluster labels
labels = kmeans.labels_

# reshape the image to original shape
clustered_image = labels.reshape(top_half.shape[0], top_half.shape[1])

# display clustered image
plt.imshow(clustered_image)
plt.show()
```

```python
corner_clusters = [clustered_image[0, 0], clustered_image[0, -1], clustered_image[-1, 0], clustered_image[-1, -1]]
non_player_clusters = max(set(corner_clusters), key=corner_clusters.count)
print(non_player_clusters)
```
